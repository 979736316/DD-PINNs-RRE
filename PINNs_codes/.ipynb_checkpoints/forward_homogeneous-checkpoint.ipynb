```python
class neural_net(object):
    def __init__(self, *inputs, layers, monotonic = False, normalize = False, LAA = False):

        self.layers = layers
        self.num_layers = len(self.layers)
        self.monotonic = monotonic
        self.normalize = normalize
        self.LAA = LAA
        
        if self.normalize:
            if len(inputs) == 0:
                in_dim = self.layers[0]
                self.X_mean = np.zeros([1, in_dim])
                self.X_std = np.ones([1, in_dim])
            else:
                X = np.concatenate(inputs, 1)
                self.X_mean = X.mean(0, keepdims=True)
                self.X_std = X.std(0, keepdims=True)

        self.weights = []
        self.biases = []
        self.A = []

        for l in range(0,self.num_layers-1):
            in_dim = self.layers[l]
            out_dim = self.layers[l+1]
            xavier_stddev = np.sqrt(2/(in_dim + out_dim))
            W = tf.Variable(tf.random.truncated_normal([in_dim, out_dim], stddev = xavier_stddev),dtype=tf.float32, trainable=True) 
            # monotonically increasing neural network
            if self.monotonic:
                W = W**2
            b = tf.Variable(np.zeros([1, out_dim]), dtype=tf.float32, trainable=True)
            
            # tensorflow variables
            self.weights.append(W)
            self.biases.append(b)
            
            # locally adaptive activation
            if self.LAA:
                a = tf.Variable(0.05, dtype=tf.float32)
                self.A.append(a)
            
    def __call__(self, *inputs):
        if self.normalize:
            H = (tf.concat(inputs, 1) - self.X_mean)/self.X_std
        else:
            H = tf.concat(inputs, 1)

        for l in range(0, self.num_layers-1):
            W = self.weights[l]
            b = self.biases[l]
            
            # matrix multiplication
            H = tf.matmul(H, W)
            # add bias
            H = H + b
            # activation
            if l < self.num_layers-2:
                if self.LAA:
                    H = tf.tanh(20 * self.A[l]*H)
                else:
                    H = tf.tanh(H)

        return H
```

```python
def tf_session():
    # tf session
    config = tf.ConfigProto(allow_soft_placement=True,  # an operation might be placed on CPU for example if GPU is not available
                            log_device_placement=True)  # print out device information
    config.gpu_options.allow_growth = True
    sess = tf.Session(config=config)

    # init
    init = tf.global_variables_initializer()
    sess.run(init)

    return sess
```

```python
def Gardner(psi, theta_r, theta_s, alpha, K_s):
    K = K_s * tf.exp(alpha * psi)
    theta = theta_r + (theta_s - theta_r) * tf.exp(alpha * psi)
    return theta, K
```

```python
def log10(x):
    numerator = tf.log(x)
    denominator = tf.log(tf.constant(10, dtype=numerator.dtype))
    return numerator / denominator
```

```python
def RRE_mixed_1D(psi, theta, K, t, z):
    
    theta_t = tf.gradients(theta, t)[0]
    theta_z = tf.gradients(theta, z)[0]
    
    psi_z = tf.gradients(psi, z)[0]
    psi_zz = tf.gradients(psi_z, z)[0]
    
    K_z = tf.gradients(K, z)[0]

    qz = -K*(psi_z + 1.0) # positive upward
    
    residual = theta_t - K_z*psi_z - K*psi_zz - K_z
    
    return residual, qz
```

```python
class Sampler:
    # Initialize the class
    def __init__(self, dim, coords, func, name = None):
        self.dim = dim
        self.coords = coords
        self.func = func
        self.name = name
    def sample(self, N):
        x = self.coords[0:1,:] + (self.coords[1:2,:]-self.coords[0:1,:])*np.random.rand(N, self.dim)
        y = self.func(x)
        return x, y
```

```python
class PINNs_RRE1D(object):
    """
    PINNs_RRE1D class implements physics-informed neural network solver for 
    the Ricahrdson-Ricahrds equation for a homogeneous soil using TensorFlow 1.15.
    # notational conventions
    _tf: placeholders
    _pred: output of neural networks or other parametric models
    _res: values related to residual points
    _data: training data 
    _ic: initial condition
    _bc: boundary condition (ub: upper; lb: lower for 1D)
    _star: prediction (used after training)
    t: time
    z: vertical coordinate (positive upward)
    theta: volumetric water content
    psi: matric potential
    K: hydraulic conductivity
    ALR: adaptive learning rate
    LAA: layer-wise locally adaptive activation functions
    weight_values: [ic, ub, lb]
    """
    def __init__(self, t_res, z_res,
                       t_ic, z_ic, theta_ic,
                       t_ub, z_ub, q_ub,
                       t_lb, z_lb, theta_lb,
                       layers_psi, hydraulic_model, hydraulic_parameters, weight_values,
                       normalize, ALR, LAA):
        
        # NN architecture
        self.layers_psi = layers_psi
        
        # mode 
        self.hydraulic_model = hydraulic_model  # "monotonic", "VGM", "Brooks", "PDI", "Gardner"
        self.normalize = normalize # normalize input values to neural networks
        self.ALR = ALR  # adaptive learning rate
        self.LAA = LAA  # locally adaptive activation functions
        
        # adaptive re-weighting constant
        self.rate = 0.9
        
        # parameter for the output of neural networks
        self.beta = 1.0
        
        # initial values for weights in the loss function
        self.adaptive_constant_ic_val = np.array(weight_values[0])
        self.adaptive_constant_ub_val = np.array(weight_values[1])
        self.adaptive_constant_lb_val = np.array(weight_values[2])
        
        # hydraulic parameters  
                    
        if self.hydraulic_model == "Gardner":  # theta_r, theta_s, alpha, K_s
            self.theta_r = tf.Variable(hydraulic_parameters[0], dtype=tf.float32, trainable=False)
            self.theta_s = tf.Variable(hydraulic_parameters[1], dtype=tf.float32, trainable=False)
            self.alpha = tf.Variable(hydraulic_parameters[2], dtype=tf.float32, trainable=False)
            self.K_s = tf.Variable(hydraulic_parameters[3], dtype=tf.float32, trainable=False)

        # data
        
        [self.t_res, self.z_res] = [t_res, z_res]
        [self.t_ic, self.z_ic, self.theta_ic] = [t_ic, z_ic, theta_ic]
        [self.t_ub, self.z_ub, self.qz_ub] = [t_ub, z_ub, q_ub]
        [self.t_lb, self.z_lb, self.theta_lb] = [t_lb, z_lb, theta_lb]
        
        # placeholders
        
        [self.t_res_tf, self.z_res_tf] = [tf.placeholder(tf.float32, shape=[None, 1]) for _ in range(2)]
        [self.t_ic_tf, self.z_ic_tf, self.theta_ic_tf] = [tf.placeholder(tf.float32, shape=[None, 1]) for _ in range(3)]
        [self.t_ub_tf, self.z_ub_tf, self.qz_ub_tf] = [tf.placeholder(tf.float32, shape=[None, 1]) for _ in range(3)]
        [self.t_lb_tf, self.z_lb_tf, self.theta_lb_tf] = [tf.placeholder(tf.float32, shape=[None, 1]) for _ in range(3)]

        self.adaptive_constant_ic_tf = tf.placeholder(tf.float32, shape=self.adaptive_constant_ic_val.shape)
        self.adaptive_constant_ub_tf = tf.placeholder(tf.float32, shape=self.adaptive_constant_ub_val.shape)
        self.adaptive_constant_lb_tf = tf.placeholder(tf.float32, shape=self.adaptive_constant_lb_val.shape)
        
        # neural network definition
        self.net_psi = neural_net(self.t_res, self.z_res, layers = self.layers_psi, monotonic = False, normalize = self.normalize, LAA = self.LAA)
        
        # prediction at initial condition
        self.log_psi_ic_pred = self.net_psi(self.t_ic_tf, self.z_ic_tf)
        self.psi_ic_pred = -tf.exp(self.log_psi_ic_pred) + self.beta
        self.theta_ic_pred, self.K_ic_pred = Gardner(self.psi_ic_pred, self.theta_r, self.theta_s, self.alpha, self.K_s)
        
        # prediction at upper boundary
        self.log_psi_ub_pred = self.net_psi(self.t_ub_tf, self.z_ub_tf)
        self.psi_ub_pred = -tf.exp(self.log_psi_ub_pred) + self.beta
        self.theta_ub_pred, self.K_ub_pred = Gardner(self.psi_ub_pred, self.theta_r, self.theta_s, self.alpha, self.K_s)
        self.qz_ub_pred = RRE_mixed_1D(self.psi_ub_pred, self.theta_ub_pred, self.K_ub_pred,
                                                                self.t_ub_tf, self.z_ub_tf)[1]
        # prediction at lower boundary
        self.log_psi_lb_pred = self.net_psi(self.t_lb_tf, self.z_lb_tf)
        self.psi_lb_pred = -tf.exp(self.log_psi_lb_pred) + self.beta
        self.theta_lb_pred, self.K_lb_pred = Gardner(self.psi_lb_pred, self.theta_r, self.theta_s, self.alpha, self.K_s)
                    
        # prediction at residual points
        self.log_psi_res_pred = self.net_psi(self.t_res_tf, self.z_res_tf)
        self.psi_res_pred = -tf.exp(self.log_psi_res_pred) + self.beta 
        self.theta_res_pred, self.K_res_pred = Gardner(self.psi_res_pred, self.theta_r, self.theta_s, self.alpha, self.K_s)
              
        self.residual_res_pred, self.qz_res_pred = RRE_mixed_1D(self.psi_res_pred, self.theta_res_pred, self.K_res_pred,
                                                                self.t_res_tf, self.z_res_tf)
        
        # loss
        
        self.loss_res =  tf.reduce_mean(tf.square(self.residual_res_pred))
        self.loss_ic = tf.reduce_mean(tf.square(self.theta_ic_pred - self.theta_ic_tf))
        self.loss_ub = tf.reduce_mean(tf.square(self.qz_ub_pred - self.qz_ub_tf))
        self.loss_lb = tf.reduce_mean(tf.square(self.theta_lb_pred - self.theta_lb_tf))
        self.loss = self.loss_res +  self.adaptive_constant_ic_tf * self.loss_ic   \
                 +  self.adaptive_constant_ub_tf * self.loss_ub \
                 +  self.adaptive_constant_lb_tf * self.loss_lb
                    
        # define optimizer with learning rate schedule
        self.global_step = tf.Variable(0, trainable = False)
        self.starter_learning_rate = 1e-3
        self.learning_rate = tf.train.exponential_decay(self.starter_learning_rate, self.global_step,
                                                        1000, 0.90, staircase=False)
        
        self.train_op = tf.train.AdamOptimizer(self.learning_rate).minimize(self.loss, global_step=self.global_step)

        # L-BFGS-B method
        self.optimizer = tf.contrib.opt.ScipyOptimizerInterface(self.loss,
                                                        method = 'L-BFGS-B',
                                                        options = {'maxiter': 50000,
                                                                   'maxfun': 50000,
                                                                   'maxcor': 50,
                                                                   'maxls': 50,
                                                                   'ftol' : 1e-10,
                                                                   'gtol' : 1e-8})
        
        # logger
        
        self.loss_res_log = []
        self.loss_ic_log = []
        self.loss_ub_log = []
        self.loss_lb_log = []
        
        self.loss_res_log_BFGS = []
        self.loss_ic_log_BFGS = []
        self.loss_ub_log_BFGS = []
        self.loss_lb_log_BFGS = []
        
        self.adaptive_activation = []

        self.saver = tf.train.Saver()
        
        # new residual points
        self.new_residual = []
        
        # generate dicts for gradients storage
        
        self.dict_gradients_res_layers = self.generate_grad_dict(self.layers_psi)
        self.dict_gradients_ic_layers = self.generate_grad_dict(self.layers_psi)
        self.dict_gradients_ub_layers = self.generate_grad_dict(self.layers_psi)
        self.dict_gradients_lb_layers = self.generate_grad_dict(self.layers_psi)
        
        # gradients Storage
        
        self.grad_res = []
        self.grad_ic = []
        self.grad_ub = []
        self.grad_lb = []
        
        for i in range(len(self.layers_psi) - 1):
            
            self.grad_res.append(tf.gradients(self.loss_res, self.net_psi.weights[i])[0])
            self.grad_ic.append(tf.gradients(self.adaptive_constant_ic_tf * self.loss_ic, self.net_psi.weights[i])[0])
            self.grad_ub.append(tf.gradients(self.adaptive_constant_ub_tf * self.loss_ub, self.net_psi.weights[i])[0])
            self.grad_lb.append(tf.gradients(self.adaptive_constant_lb_tf * self.loss_lb, self.net_psi.weights[i])[0])
        
        # store the adaptive constant
        
        self.adaptive_constant_ic_log = []
        self.adaptive_constant_ub_log = []
        self.adaptive_constant_lb_log = []
        
        # compute the adaptive constant
        
        self.adaptive_constant_ic_list = []
        self.adaptive_constant_ub_list = []
        self.adaptive_constant_lb_list = []
        
        self.max_grad_res_list = []
        
        self.mean_grad_ic_list = []
        self.mean_grad_ub_list = []
        self.mean_grad_lb_list = []
        
        for i in range(len(self.layers_psi) - 1):
            self.max_grad_res_list.append(tf.reduce_max(tf.abs(self.grad_res[i]))) 
            
            self.mean_grad_ic_list.append(tf.reduce_mean(tf.abs(self.grad_ic[i])))
            self.mean_grad_ub_list.append(tf.reduce_mean(tf.abs(self.grad_ub[i])))
            self.mean_grad_lb_list.append(tf.reduce_mean(tf.abs(self.grad_lb[i])))
            
        self.max_grad_res = tf.reduce_max(tf.stack(self.max_grad_res_list))
        
        self.mean_grad_ic = tf.reduce_mean(tf.stack(self.mean_grad_ic_list))
        self.mean_grad_ub = tf.reduce_mean(tf.stack(self.mean_grad_ub_list))
        self.mean_grad_lb = tf.reduce_mean(tf.stack(self.mean_grad_lb_list))
       
        
        self.adaptive_constant_ic = self.max_grad_res / self.mean_grad_ic
        self.adaptive_constant_ub = self.max_grad_res / self.mean_grad_ub
        self.adaptive_constant_lb = self.max_grad_res / self.mean_grad_lb
        
        self.sess = tf_session()
    
    # create dictionary to store gradients
    def generate_grad_dict(self, layers):
        num = len(layers) - 1
        grad_dict = {}
        for i in range(num):
            grad_dict['layer_{}'.format(i + 1)] = []
        return grad_dict
    
    # save gradients
    def save_gradients(self, tf_dict):
        num_layers = len(self.layers_psi)
        for i in range(num_layers - 1):
            grad_res_value , grad_ic_value, grad_ub_value, grad_lb_value = self.sess.run([self.grad_res[i],
                                                               self.grad_ic[i],
                                                               self.grad_ub[i],
                                                               self.grad_lb[i]],
                                                               feed_dict=tf_dict)

            # save gradients of loss_res and loss_bcs
            self.dict_gradients_res_layers['layer_' + str(i + 1)].append(grad_res_value.flatten())
            
            self.dict_gradients_ic_layers['layer_' + str(i + 1)].append(grad_ic_value.flatten())
            self.dict_gradients_ub_layers['layer_' + str(i + 1)].append(grad_ub_value.flatten())
            self.dict_gradients_lb_layers['layer_' + str(i + 1)].append(grad_lb_value.flatten())
            
        return None
    
    def train_Adam(self, iterations, batch = True, batch_size = 128):

        N_res = self.t_res.shape[0]
        

        start_time = timeit.default_timer()
        running_time = 0
        it = 0
        for it in range(iterations):
           
            if batch:

                idx_res = np.random.choice(N_res, batch_size, replace = False)

                (t_res, z_res) = (self.t_res[idx_res,:],
                                  self.z_res[idx_res,:])
            else:

                (t_res, z_res) = (self.t_res,
                                  self.z_res)

            tf_dict = {self.t_res_tf: t_res,
                       self.z_res_tf: z_res,
                       self.t_ic_tf: self.t_ic,
                       self.z_ic_tf: self.z_ic,
                       self.theta_ic_tf: self.theta_ic,
                       self.t_ub_tf: self.t_ub,
                       self.z_ub_tf: self.z_ub,
                       self.qz_ub_tf: self.qz_ub,
                       self.t_lb_tf: self.t_lb,
                       self.z_lb_tf: self.z_lb,
                       self.theta_lb_tf: self.theta_lb,
                       self.adaptive_constant_ic_tf: self.adaptive_constant_ic_val,
                       self.adaptive_constant_ub_tf: self.adaptive_constant_ub_val,
                       self.adaptive_constant_lb_tf: self.adaptive_constant_lb_val}

            self.sess.run(self.train_op, tf_dict)

            # print
            if it % 10 == 0:
                elapsed = timeit.default_timer() - start_time
                running_time += elapsed/3600.0
                loss_value = self.sess.run(self.loss, tf_dict)
                loss_res_value, loss_ic_value, loss_ub_value, loss_lb_value = self.sess.run([self.loss_res, self.loss_ic, self.loss_ub, self.loss_lb], tf_dict)
                
                adaptive_activation = self.sess.run(self.net_psi.A, tf_dict)
                #compute and print adaptive weights during training
                if self.ALR:
                    # compute the adaptive constant
                    
                    adaptive_constant_ic_val = self.sess.run(self.adaptive_constant_ic, tf_dict)
                    adaptive_constant_ub_val = self.sess.run(self.adaptive_constant_ub, tf_dict)
                    adaptive_constant_lb_val = self.sess.run(self.adaptive_constant_lb, tf_dict)
                    
                    self.adaptive_constant_ic_val = adaptive_constant_ic_val * (
                            1.0 - self.rate) + self.rate * self.adaptive_constant_ic_val
                    self.adaptive_constant_ub_val = adaptive_constant_ub_val * (
                            1.0 - self.rate) + self.rate * self.adaptive_constant_ub_val
                    self.adaptive_constant_lb_val = adaptive_constant_lb_val * (
                            1.0 - self.rate) + self.rate * self.adaptive_constant_lb_val
                    
            # store loss and adaptive weights
                
                self.loss_res_log.append(loss_res_value)
                self.loss_ic_log.append(loss_ic_value)
                self.loss_ub_log.append(loss_ub_value)
                self.loss_lb_log.append(loss_lb_value)
                self.adaptive_activation.append(adaptive_activation)
                
                self.adaptive_constant_ic_log.append(self.adaptive_constant_ic_val)
                self.adaptive_constant_ub_log.append(self.adaptive_constant_ub_val)
                self.adaptive_constant_lb_log.append(self.adaptive_constant_lb_val)
                
                print('It: %d, Loss: %.3e, Loss_r: %.3e, Loss_ic: %.3e, Loss_ub: %.3e, Loss_lb: %.3e, Time: %.2f' %
                      (it, loss_value, loss_res_value, loss_ic_value, loss_ub_value, loss_lb_value, elapsed))
                print("constant_ic_val: {:.3f}, constant_ub_val: {:.3f}, constant_lb_val: {:.3f}".format(
                    self.adaptive_constant_ic_val, self.adaptive_constant_ub_val, self.adaptive_constant_lb_val))
        
                start_time = timeit.default_timer()
            
            # store gradients
            if it == iterations - 1:  # at the end of the training (if adaptive residual refinement is chosen, the end of every Adam)
                self.save_gradients(tf_dict)
                print("Gradients information stored ...")                                  
    
    def train_BFGS(self):
        # full batch L-BFGS
        tf_dict = {
           self.t_res_tf: self.t_res,
           self.z_res_tf: self.z_res,
           self.t_ic_tf: self.t_ic,
           self.z_ic_tf: self.z_ic,
           self.theta_ic_tf: self.theta_ic,
           self.t_ub_tf: self.t_ub,
           self.z_ub_tf: self.z_ub,
           self.qz_ub_tf: self.qz_ub,
           self.t_lb_tf: self.t_lb,
           self.z_lb_tf: self.z_lb,
           self.theta_lb_tf: self.theta_lb,
           self.adaptive_constant_ic_tf: self.adaptive_constant_ic_val,
           self.adaptive_constant_ub_tf: self.adaptive_constant_ub_val,
           self.adaptive_constant_lb_tf: self.adaptive_constant_lb_val}

#         L-BFGS-B
        self.optimizer.minimize(self.sess,
                                feed_dict = tf_dict,
                                fetches = [self.loss, self.loss_res, self.loss_ic, self.loss_ub, self.loss_lb],
                                loss_callback = self.callback)
        
    def callback(self, loss_value, loss_res_value, loss_ic_value, loss_ub_value, loss_lb_value):
        
        tf_dict = {
           self.t_res_tf: self.t_res,
           self.z_res_tf: self.z_res,
           self.t_ic_tf: self.t_ic,
           self.z_ic_tf: self.z_ic,
           self.theta_ic_tf: self.theta_ic,
           self.t_ub_tf: self.t_ub,
           self.z_ub_tf: self.z_ub,
           self.qz_ub_tf: self.qz_ub,
           self.t_lb_tf: self.t_lb,
           self.z_lb_tf: self.z_lb,
           self.theta_lb_tf: self.theta_lb,
           self.adaptive_constant_ic_tf: self.adaptive_constant_ic_val,
           self.adaptive_constant_ub_tf: self.adaptive_constant_ub_val,
           self.adaptive_constant_lb_tf: self.adaptive_constant_lb_val}
        
        self.loss_res_log_BFGS.append(loss_res_value)
        self.loss_ic_log_BFGS.append(loss_ic_value)
        self.loss_ub_log_BFGS.append(loss_ub_value)
        self.loss_lb_log_BFGS.append(loss_lb_value)
        
        
        print('Loss: %.3e, Loss_r: %.3e, Loss_ic: %.3e, Loss_ub: %.3e, Loss_lb: %.3e' %
                      (loss_value, loss_res_value, loss_ic_value, loss_ub_value, loss_lb_value))

        
    def RAR(self, dom_coords, iterations, batchsize, m):
        self.train_Adam(iterations, batchsize)
        
        i = 1
        res_sampler = Sampler(2, dom_coords, lambda x: 0)
        Z_res, _ = res_sampler.sample(np.int32(1e6))
        t_res, z_res = Z_res[:, 0:1], Z_res[:, 1:2]
    
        tf_dict = {self.t_res_tf: t_res, self.z_res_tf: z_res}
        
        while i < 10:
            
            residual = self.sess.run(self.residual_res_pred, tf_dict)
            err_abs = np.abs(residual)
            err = np.mean(err_abs)
            print("Mean residual: %.3e" % (err))

            # choose m largest residual
            z_id = err_abs.argsort(axis = 0)[-m:][::-1][:,0]

            print(z_id)
            print("Adding new point:", Z_res[z_id, :], "\n")
            print("High residuals are:", err_abs[z_id], "\n")

            self.new_residual.append(Z_res[z_id, :])
            self.t_res = np.vstack((self.t_res, Z_res[z_id, 0:1]))
            self.z_res = np.vstack((self.z_res, Z_res[z_id, 1:2]))
            
            self.train_Adam(iterations, batchsize)
            
            i += 1

            print(i)
            
    def predict(self, t_star, z_star):

        tf_dict = {self.t_res_tf: t_star, self.z_res_tf: z_star}

        psi_star = self.sess.run(self.psi_res_pred, tf_dict)
        theta_star = self.sess.run(self.theta_res_pred, tf_dict)
        K_star = self.sess.run(self.K_res_pred, tf_dict)
        qz_star = self.sess.run(self.qz_res_pred, tf_dict)
        
        residual = self.sess.run(self.residual_res_pred, tf_dict)

        return psi_star, theta_star, K_star, qz_star, residual
```

```python
def main_loop(folder, ID, random_seed, iterations, number_res = 10000, RAR = False, LAA = True, number_layers = 5, number_units = 50,    
              number_ub = 1000, 
              batch_size = 128, weight_values = [10, 10, 10], ALR = False, new_residual_points = 10):
    """
    folder: the name of the folder to save
    ALR: adaptive learning rate
    LAA: layer-wise locally adaptive activation functions
    RAR: residual adaptive refinement
    weight_values: [ic, ub, lb]
    """
    if not os.path.isdir(f'../results/{folder}'):
        os.mkdir(f'../results/{folder}')
        
    if not os.path.isdir(f'../results/{folder}/{ID}'):
        os.mkdir(f'../results/{folder}/{ID}')
    
    # domain boundaries; the first column is time, the second column is depth (positive upward)
    ics_coords = np.array([[0.0, 0.0],
                           [0.0, 10.0]])
    lb_coords = np.array([[0.0, 0.0],   # upper boundary condition
                           [10.0, 0.0]])
    ub_coords = np.array([[0.0, 10.0],   # lower boundary condition
                           [10.0, 10.0]])
    dom_coords = np.array([[0.0, 0.0],   # t = 0 hour to t = 10 hours
                           [10.0, 10.0]])

    # import true solution: first index is depth, second index is time
    psi_true = np.load(f"../analytical_solutions/Srivastava_psi_homogeneous.npy")

    # import residual points 
    Z_res = np.load(f"data/residual_points.npy")

    # hydraulic paramters
    hydraulic_parameters = Srivastava_parameters["homogeneous"]

    # hydralic function model
    hydraulic_model = "Gardner"

    # normalization
    normalize = False

    # NN architecture
    layers_psi = [2] + number_layers*[number_units] + [1]

    # forward solution
    # collocation points
    if RAR:
        number_res2 =  number_res - (10 - 1) * new_residual_points
    else:
        number_res2 = number_res

    t_res = Z_res[:number_res2, 0:1]
    z_res = Z_res[:number_res2, 1:2]

    # initial points

    t_ic = np.linspace(ics_coords[0, 0], ics_coords[1, 0], 101)[:, None]
    z_ic = np.linspace(ics_coords[0, 1], ics_coords[1, 1], 101)[:, None]

    # upper boundary points

    t_ub = np.linspace(ub_coords[0, 0], ub_coords[1, 0], number_ub + 1)[1:, None]
    z_ub = np.linspace(ub_coords[0, 1], ub_coords[1, 1], number_ub + 1)[1:, None]

    # lower boundary points

    t_lb = np.linspace(lb_coords[0, 0], lb_coords[1, 0], 101)[:, None]
    z_lb = np.linspace(lb_coords[0, 1], lb_coords[1, 1], 101)[:, None]

    # initial condition
    # upper water flux condition (positive downward) [cm/hour]
    q_A_star = 0.1
    q_B_star = 0.9

    # parameters
    K_s = 1.0
    alpha = 1.0
    psi_0 = 0.0
    theta_r = 0.06
    theta_s = 0.40

    q_A = q_A_star / K_s
    q_B = q_B_star / K_s

    K_ic = q_A - (q_A - np.exp(alpha * psi_0)) * np.exp(-z_ic * alpha)
    psi_ic = np.log(K_ic) /alpha

    theta_ic = theta_r + (theta_s - theta_r) * np.exp(alpha * psi_ic)

    # boundary conditions
    theta_lb = theta_r + (theta_s - theta_r) * np.exp(alpha * psi_0)
    qz_ub = -q_B_star * np.ones(t_ub.shape)
    theta_lb = theta_lb * np.ones(t_lb.shape)

    # initialization
    tf.reset_default_graph()
    tf.set_random_seed(random_seed)
    random.seed(random_seed)
    np.random.seed(random_seed)

    # training    
    model = PINNs_RRE1D(
                t_res, z_res,
                t_ic, z_ic, theta_ic,
                t_ub, z_ub, qz_ub,
                t_lb, z_lb, theta_lb,    
                layers_psi, hydraulic_model, hydraulic_parameters, weight_values, normalize, ALR, LAA
                   )

    # Adam training
    start_time = timeit.default_timer() 

    if RAR:
        iterations = iterations//10
        model.RAR(dom_coords, iterations, batch_size, new_residual_points)
    else:
        model.train_Adam(iterations, batch_size)
    time_Adam = timeit.default_timer() - start_time

    # BFGS training
    start_time = timeit.default_timer() 
    model.train_BFGS()
    time_BFGS = timeit.default_timer() - start_time

    # evaluation
    T = 101
    N = 101
    t = np.linspace(dom_coords[0, 0], dom_coords[1, 0], T)[:, None]
    z = np.linspace(dom_coords[0, 1], dom_coords[1, 1], N)[:, None]
    z_mesh, t_mesh = np.meshgrid(z, t)
    X_star = np.hstack((t_mesh.flatten()[:, None], z_mesh.flatten()[:, None]))

    t_test = X_star[:, 0:1]
    z_test = X_star[:, 1:2]

    # prediction
    psi_pred, theta_pred, K_pred, flux_pred, residual = model.predict(t_test, z_test)

    theta_pred = theta_pred.reshape((T, N))
    K_pred = K_pred.reshape((T, N))
    psi_pred = psi_pred.reshape((T, N))
    flux_pred = flux_pred.reshape((T, N))
    residual = residual.reshape((T, N))

    theta_true = theta_r + (theta_s - theta_r) * np.exp(alpha * psi_true)

    theta_L2_error = np.sqrt(np.mean((theta_true.T - theta_pred) ** 2))
    theta_abs_error = np.mean(np.abs(theta_true.T- theta_pred))
    theta_square = np.sqrt(np.mean((theta_true.T) ** 2))
    theta_max_error = np.max(np.abs(theta_true.T- theta_pred))

    theta_L2_relative_error = theta_L2_error/theta_square
    theta_abs_relative_error = theta_abs_error/np.mean(np.abs(theta_true.T))

    print("theta_L2_relative_error is", theta_L2_relative_error)
    print("theta_abs_relative_error is", theta_abs_relative_error)
    print("theta_max_error is", theta_max_error)

    psi_L2_error = np.sqrt(np.mean((psi_true.T - psi_pred) ** 2))
    psi_abs_error = np.mean(np.abs(psi_true.T- psi_pred))
    psi_square = np.sqrt(np.mean((psi_true.T) ** 2))
    psi_max_error = np.max(np.abs(psi_true.T- psi_pred))

    psi_L2_relative_error = psi_L2_error/psi_square
    psi_abs_relative_error = psi_abs_error/np.mean(np.abs(psi_true.T))


    print("psi_L2_relative_error is", psi_L2_relative_error)
    print("psi_abs_relative_error is", psi_abs_relative_error)
    print("psi_max_error is", psi_max_error)

    # theta distribution
    fig, ax = plt.subplots(1, 3, figsize=(18, 5))

    c1 = ax[0].pcolor(t_mesh, z_mesh - 10.0, theta_pred, cmap='gnuplot2_r')
    fig.colorbar(c1, ax = ax[0])
    ax[0].set_xlabel(r'$t$ [hours]')
    ax[0].set_ylabel(r'$z$ [cm]')
    ax[0].set_title('Predicted $\\theta$')

    c2 = ax[1].pcolor(t_mesh, z_mesh - 10.0, theta_true.T, cmap='gnuplot2_r')
    fig.colorbar(c2, ax = ax[1])
    ax[1].set_xlabel(r'$t$ [hours]')
    ax[1].set_title('True $\\theta$')

    c3 = ax[2].pcolor(t_mesh, z_mesh - 10.0, np.abs(theta_pred - theta_true.T), cmap='gnuplot2_r')
    fig.colorbar(c3, ax = ax[2])
    ax[2].set_xlabel(r'$t$ [hours]')
    ax[2].set_title('Absolute error in $\\theta$')

    fig.savefig(f'../results/{folder}/{ID}/theta_color.png')
    fig.clf()


    # psi distribution
    fig, ax = plt.subplots(1, 3, figsize=(18, 5))

    c1 = ax[0].pcolor(t_mesh, z_mesh - 10.0, psi_pred, cmap='gnuplot2_r')
    fig.colorbar(c1, ax = ax[0])
    ax[0].set_xlabel(r'$t$ [hours]')
    ax[0].set_ylabel(r'$z$ [cm]')
    ax[0].set_title('Predicted $\\psi$ [cm]')

    c2 = ax[1].pcolor(t_mesh, z_mesh - 10.0, psi_true.T, cmap='gnuplot2_r')
    fig.colorbar(c2, ax = ax[1])
    ax[1].set_xlabel(r'$t$ [hours]')
    ax[1].set_title('True $\\psi$ [cm]')

    c3 = ax[2].pcolor(t_mesh, z_mesh - 10.0, np.abs(psi_pred - psi_true.T), cmap='gnuplot2_r')
    fig.colorbar(c3, ax = ax[2])
    ax[2].set_xlabel(r'$t$ [hours]')
    ax[2].set_title('Absolute error in $\\psi$ [cm]')

    fig.savefig(f'../results/{folder}/{ID}/psi_color.png')
    fig.clf()

    fig, ax = plt.subplots(1, 1, figsize=(6.5, 6.5))

    c1 = ax.pcolor(t_mesh, z_mesh - 10.0, flux_pred, cmap='gnuplot2_r')
    fig.colorbar(c1, ax = ax)
    ax.set_xlabel(r'$t$ [hours]')
    ax.set_ylabel(r'$z$ [cm]')
    ax.set_title('Predicted $q$ [cm/hours]')

    fig.savefig(f'../results/{folder}/{ID}/flux.png')
    fig.clf()

    fig = plt.subplots(1, 1, figsize=(6.5, 6.5))

    plt.scatter(theta_pred[0, ::5], z_mesh[0, ::5] - 10.0, s = 15, label = 'Predicted')
    plt.scatter(theta_pred[1, ::5], z_mesh[0, ::5] - 10.0, s = 15)
    plt.scatter(theta_pred[5, ::5], z_mesh[0, ::5] - 10.0, s = 15)
    plt.scatter(theta_pred[10, ::5], z_mesh[0, ::5] - 10.0, s = 15)
    plt.scatter(theta_pred[30, ::5], z_mesh[0, ::5] - 10.0, s = 15)
    plt.scatter(theta_pred[50, ::5], z_mesh[0, ::5] - 10.0, s = 15)
    plt.scatter(theta_pred[100, ::5], z_mesh[0, ::5] - 10.0, s = 15)

    plt.plot(theta_true.T[0, :], z_mesh[0, :] - 10.0 ,label = 't = 0 hour')
    plt.plot(theta_true.T[1, :], z_mesh[0, :] - 10.0,label = 't = 0.1 hour')
    plt.plot(theta_true.T[5, :], z_mesh[0, :] - 10.0,label = 't = 0.5 hour')
    plt.plot(theta_true.T[10, :], z_mesh[0, :]- 10.0 ,label = 't = 1 hour')
    plt.plot(theta_true.T[30, :], z_mesh[0, :]- 10.0,label = 't = 3 hour')
    plt.plot(theta_true.T[50, :], z_mesh[0, :]- 10.0,label = 't = 5 hour')
    plt.plot(theta_true.T[100, :], z_mesh[0, :]- 10.0, label = 't = 10 hour')

    plt.xlim(0.05,0.50)
    plt.xlabel('$\\theta$')
    plt.ylabel('$z$ cm')
    plt.legend(loc = (0.75,0.55), fontsize = 8)

    plt.savefig(f'../results/{folder}/{ID}/theta_fixed_time.png')
    plt.clf()

    fig = plt.figure(1, figsize=(6.5, 6.5))

    plt.plot(t_res, z_res - 10.0, 'rx', markersize = 3, label = "residual points")

    if RAR:
        new_residual = model.new_residual[0]
        for i in range(1, len(model.new_residual)):
            new_residual = np.vstack((new_residual, model.new_residual[i]))
        plt.plot(new_residual[:, 0], new_residual[:, 1] - 10.0, 'bx', markersize = 5, label = "added residual points")



    plt.xlabel(r'$t$ [hours]')
    plt.ylabel(r'$z$ [cm]')
    plt.title('residual points')
    plt.xlim(-0.1, 10.1)
    plt.ylim(-10.1, 0.1)
    plt.legend(loc = (0.55,0.15))
    plt.savefig(f'../results/{folder}/{ID}/residual_points.png', transparent = True)
    plt.clf()


    fig = plt.figure(1, figsize=(6.5, 6.5))

    plt.pcolor(t_mesh, z_mesh - 10.0, np.log10(np.abs(residual)), cmap='gnuplot2_r')

    plt.colorbar()
    plt.xlabel(r'$t$ [hours]')
    plt.ylabel(r'$z$ [cm]')
    plt.title('$log_{10}r(x)$')
    plt.savefig(f'../results/{folder}/{ID}/residual.png', transparent = True)
    plt.clf()

    # Loss

    loss_ub = model.loss_ub_log
    loss_lb = model.loss_lb_log
    loss_ic = model.loss_ic_log
    loss_res = model.loss_res_log

    Adam_iterations = np.arange(0, len(loss_res)) * 10

    loss_ub_BFGS = model.loss_ub_log_BFGS
    loss_lb_BFGS = model.loss_lb_log_BFGS
    loss_ic_BFGS = model.loss_ic_log_BFGS
    loss_res_BFGS = model.loss_res_log_BFGS

    BFGS_iterations = np.arange(0, len(loss_res_BFGS))

    total_iterations = np.append(Adam_iterations, BFGS_iterations + Adam_iterations[-1])

    fig, ax = plt.subplots(1, 3, figsize=(18, 5))

    # Adam
    ax[0].plot(Adam_iterations, loss_ic, label='$\mathcal{L}_{ic}$')
    ax[0].plot(Adam_iterations, loss_ub ,label='$\mathcal{L}_{ub}$')
    ax[0].plot(Adam_iterations, loss_lb,label='$\mathcal{L}_{lb}$')
    ax[0].plot(Adam_iterations, loss_res, label='$\mathcal{L}_{res}$')
    ax[0].set_yscale('log')
    ax[0].set_xlabel('iterations')
    ax[0].set_ylabel('Loss')
    ax[0].set_title('Adam')
    ax[0].legend()

    # BFGS
    ax[1].plot(BFGS_iterations, loss_ic_BFGS, label='$\mathcal{L}_{ic}$')
    ax[1].plot(BFGS_iterations, loss_ub_BFGS, label='$\mathcal{L}_{ub}$')
    ax[1].plot(BFGS_iterations, loss_lb_BFGS, label='$\mathcal{L}_{lb}$')
    ax[1].plot(BFGS_iterations, loss_res_BFGS, label='$\mathcal{L}_{res}$')
    ax[1].set_yscale('log')
    ax[1].set_xlabel('iterations')
    ax[1].set_ylabel('Loss')
    ax[1].set_title('L-BFGS-B')
    ax[1].legend()


    # total training

    ax[2].plot(total_iterations, loss_ic + loss_ic_BFGS, label='$\mathcal{L}_{ic}$')
    ax[2].plot(total_iterations, loss_ub + loss_ub_BFGS ,label='$\mathcal{L}_{ub}$')
    ax[2].plot(total_iterations, loss_lb + loss_lb_BFGS,label='$\mathcal{L}_{lb}$')
    ax[2].plot(total_iterations, loss_res + loss_res_BFGS, label='$\mathcal{L}_{res}$')
    ax[2].set_yscale('log')
    ax[2].set_xlabel('iterations')
    ax[2].set_ylabel('Loss')
    ax[2].set_title('total')
    ax[2].legend()

    plt.savefig(f'../results/{folder}/{ID}/training.png')
    plt.clf()

    # Adaptive Constant

    adaptive_constant_ic = model.adaptive_constant_ic_log
    adaptive_constant_ub = model.adaptive_constant_ub_log
    adaptive_constant_lb = model.adaptive_constant_lb_log

    Adam_iterations = np.arange(0, len(adaptive_constant_ic )) * 10

    fig = plt.subplots(1, 1, figsize=(6.5, 6.5))

    plt.plot(Adam_iterations, adaptive_constant_ic, label='$\lambda_{u_{ic}}$')
    plt.plot(Adam_iterations, adaptive_constant_ub, label='$\lambda_{u_{ub}}$')
    plt.plot(Adam_iterations, adaptive_constant_lb, label='$\lambda_{u_{lb}}$')
    plt.yscale('log')
    plt.xlabel('iterations')
    plt.ylabel('Weight parameters $\lambda$')
    plt.legend()

    plt.savefig(f'../results/{folder}/{ID}/dynamics_weights.png')
    plt.clf()

    # layer-wise locally adaptive parameters
    adaptive_parameter_layer1 = []
    adaptive_parameter_layer2 = []
    adaptive_parameter_layer3 = []
    adaptive_parameter_layer4 = []
    adaptive_parameter_layer5 = []
    for i in range(len(model.adaptive_activation)):
        adaptive_parameters = model.adaptive_activation[i]
        adaptive_parameter_layer1.append(adaptive_parameters[0]* 20)
        adaptive_parameter_layer2.append(adaptive_parameters[1]* 20)
        adaptive_parameter_layer3.append(adaptive_parameters[2]* 20)
        adaptive_parameter_layer4.append(adaptive_parameters[3]* 20)
        adaptive_parameter_layer5.append(adaptive_parameters[4]* 20)

    plt.plot(adaptive_parameter_layer1, label = 'layer1')
    plt.plot(adaptive_parameter_layer2, label = 'layer2')
    plt.plot(adaptive_parameter_layer3, label = 'layer3')
    plt.plot(adaptive_parameter_layer4, label = 'layer4')
    plt.plot(adaptive_parameter_layer5, label = 'layer5')
    plt.legend()
    plt.ylabel('sa')
    plt.xlabel('iteration')

    plt.savefig(f'../results/{folder}/{ID}/adaptive_activation_parameter.png', transparent = True)
    plt.clf()
    

    # gradients at the end of training
    data_gradients_res = model.dict_gradients_res_layers
    data_gradients_ic = model.dict_gradients_ic_layers
    data_gradients_ub = model.dict_gradients_ub_layers
    data_gradients_lb = model.dict_gradients_lb_layers

    num_hidden_layers = len(layers_psi) - 1
    count = 1
    fig = plt.figure(figsize=(13, 8))

    for j in range(num_hidden_layers):
        ax = plt.subplot(2, 4, count)

        gradients_res = data_gradients_res['layer_' + str(j + 1)][-1]
        gradients_ic = data_gradients_ic['layer_' + str(j + 1)][-1]
        gradients_ub = data_gradients_ub['layer_' + str(j + 1)][-1]
        gradients_lb = data_gradients_lb['layer_' + str(j + 1)][-1]

        sns.distplot(gradients_res, hist=False,
                     kde_kws={"shade": False},
                     norm_hist=True, label=r'$\nabla_\Theta \mathcal{L}_{res}$')
        sns.distplot(gradients_ic, hist=False,
                     kde_kws={"shade": False},
                     norm_hist=True, label=r'$\nabla_\Theta \lambda_{ic} \mathcal{L}_{ic}}$')
        sns.distplot(gradients_ub, hist=False,
                     kde_kws={"shade": False},
                     norm_hist=True, label=r'$\nabla_\Theta \lambda_{ub} \mathcal{L}_{ub}$')
        sns.distplot(gradients_lb, hist=False,
                     kde_kws={"shade": False},
                     norm_hist=True, label=r'$\nabla_\Theta \lambda_{lb} \mathcal{L}_{lb}$')

        ax.set_title('Layer {}'.format(j + 1))
        ax.set_yscale('symlog')

        ax.get_legend().remove()
        count += 1

    handles, labels = ax.get_legend_handles_labels()
    fig.legend(handles, labels, loc="upper left", bbox_to_anchor=(0.4, 0.1),
                 borderaxespad=0, bbox_transform=fig.transFigure, ncol=3)
    # plt.tight_layout()

    plt.savefig(f'../results/{folder}/{ID}/parameter_distributions.png', transparent = True)
    plt.clf()

    # store data

    np.save(f"../results/{folder}/{ID}/theta_pred", theta_pred)
    np.save(f"../results/{folder}/{ID}/psi_pred", psi_pred)
    np.save(f"../results/{folder}/{ID}/K_pred", K_pred)
    np.save(f"../results/{folder}/{ID}/flux_pred", flux_pred)
    np.save(f"../results/{folder}/{ID}/residual", residual)

    np.save(f"../results/{folder}/{ID}/loss_ic", loss_ic)
    np.save(f"../results/{folder}/{ID}/loss_ic_BFGS", loss_ic_BFGS)
    np.save(f"../results/{folder}/{ID}/loss_ub", loss_ub)
    np.save(f"../results/{folder}/{ID}/loss_ub_BFGS", loss_ub_BFGS)
    np.save(f"../results/{folder}/{ID}/loss_lb", loss_lb)
    np.save(f"../results/{folder}/{ID}/loss_lb_BFGS", loss_lb_BFGS)
    np.save(f"../results/{folder}/{ID}/loss_res", loss_res)
    np.save(f"../results/{folder}/{ID}/loss_res_BFGS", loss_res_BFGS)
    
    np.save(f"../results/{folder}/{ID}/adaptive_parameter_layer1", adaptive_parameter_layer1)
    np.save(f"../results/{folder}/{ID}/adaptive_parameter_layer2", adaptive_parameter_layer2)
    np.save(f"../results/{folder}/{ID}/adaptive_parameter_layer3", adaptive_parameter_layer3)
    np.save(f"../results/{folder}/{ID}/adaptive_parameter_layer4", adaptive_parameter_layer4)
    np.save(f"../results/{folder}/{ID}/adaptive_parameter_layer5", adaptive_parameter_layer5)
    
    np.save(f"../results/{folder}/{ID}/adaptive_constant_ic", adaptive_constant_ic)
    np.save(f"../results/{folder}/{ID}/adaptive_constant_ub", adaptive_constant_ub)
    np.save(f"../results/{folder}/{ID}/adaptive_constant_lb", adaptive_constant_lb)

    np.save(f"../results/{folder}/{ID}/gradients_res", data_gradients_res)
    np.save(f"../results/{folder}/{ID}/gradients_ub", data_gradients_ub)
    np.save(f"../results/{folder}/{ID}/gradients_lb", data_gradients_lb)
    np.save(f"../results/{folder}/{ID}/gradients_ic", data_gradients_ic)

    np.save(f"../results/{folder}/{ID}/t_res", t_res)
    np.save(f"../results/{folder}/{ID}/z_res", z_res)
    if RAR:
        np.save(f"../results/{folder}/{ID}/new_residual", new_residual)

    np.save(f"../results/{folder}/{ID}/z_mesh", z_mesh)
    np.save(f"../results/{folder}/{ID}/t_mesh", t_mesh)

    sheet1.write(ID, 0, ID)
    sheet1.write(ID, 1, random_seed)
    sheet1.write(ID, 2, number_layers)
    sheet1.write(ID, 3, number_units)
    sheet1.write(ID, 4, LAA)
    sheet1.write(ID, 5, number_res)
    sheet1.write(ID, 6, number_ub)
    sheet1.write(ID, 7, weight_values[0])
    sheet1.write(ID, 8, weight_values[1])
    sheet1.write(ID, 9, weight_values[2])
    sheet1.write(ID, 10, ALR)
    sheet1.write(ID, 11, iterations)
    sheet1.write(ID, 12, batch_size)
    sheet1.write(ID, 13, RAR)
    sheet1.write(ID, 14, new_residual_points)

    sheet1.write(ID, 15, time_Adam)
    sheet1.write(ID, 16, time_BFGS)
    sheet1.write(ID, 17, theta_L2_relative_error)
    sheet1.write(ID, 18, theta_abs_relative_error)
    sheet1.write(ID, 19, theta_max_error)
    sheet1.write(ID, 20, psi_L2_relative_error)
    sheet1.write(ID, 21, psi_abs_relative_error)
    sheet1.write(ID, 22, psi_max_error)
    sheet1.write(ID, 23, str(loss_ub_BFGS[-1]))
    sheet1.write(ID, 24, str(loss_lb_BFGS[-1]))
    sheet1.write(ID, 25, str(loss_ic_BFGS[-1]))
    sheet1.write(ID, 26, str(loss_res_BFGS[-1]))
    
   
```
